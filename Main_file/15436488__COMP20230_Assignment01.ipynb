## $Mathamatical$ $factorial$ $forumlas$
- #### factorial can be defined as:

$factorial(n)=n * factorial(n-1)factorial(n)=n∗factorial(n−1)$


$factorial(n)=n∗(n−1)∗(n−2)...∗1$

$for$  

$n >= 1$
#### $Note:$ 
$The$ $factorial$ $of$ $0$ $is$ $1.$

 ### $Factorial$ $algorithm:$
 
- Step 1 : Start
     - Start 2 : Read n
     - Start 3 : Initialize $counter$ variable $i$ to 1 and fact to 1
     - Start 4 : If $i$ $<=$ $n$ go to step 5 otherwise goto step 7
 - Start 5 : calculate factorial = $factorial$ $*$ $i$
 - Start 6 : Increment counter variable $i$ and goto step 4
 - Start 7 : Write factorial.
- Start 8 : Stop



```python
import matplotlib.pyplot as plt
from time import perf_counter_ns
import seaborn as sns
from scipy import stats
from scipy.optimize import curve_fit
import numpy as np
from numpy import arange
sns.set_theme(color_codes=True)
```

### $Iterative$ $function:$


```python
def it_factorial(n):
    fact = 1
    for num in range(2, n + 1):
        fact *= num
    return fact
```

Factorial is mainly used to calculate the total number of ways in which n distinct objects can be arranged into a sequence. (5 distinct objects can be arranged into a sequence in 120 ways). The iterative version uses a loop to calculate the product of all positive integers less than equal to n .

### $Recursive$ $function:$

```python
def re_factorial(n):
    if n == 0: 
        return 1
    else: 
        return re_factorial(n-1) * n

def tail_factorial(n, accumulator=1):
    if n == 0: 
        return accumulator
    else: 
        return tail_factorial(n-1, accumulator * n)
```

## $Iterative$ $approach$ $times:$

```python
n=[0,500, 1000, 1500, 2000, 2500] #same values of n for all functions
t=[]
results=[]
def itertime():
    
    for a in n:
        start =perf_counter_ns()

        it_factorial(a)
        
        #results.append(it_factorial(a))
        
        end = perf_counter_ns()

        t.append(end-start)
        
    return t,# results
    
itertime()
```

## $Recursive$ $approach$ $times$ $(non-tail):$

```python
nontail_t=[]
results_re_nontail=[]

def rec_non_tail_time():
    
    for a in n:
        start =perf_counter_ns()

        re_factorial(a)
        #results_re_nontail.append(re_factorial(a))
        
        end = perf_counter_ns()

        nontail_t.append(end-start)
        
        
    return nontail_t, #results_re_nontail
    
rec_non_tail_time()
```

## $Recursive$ $approach$ $times$ $(tail):$

```python
tail_t=[]
results_re_tail=[]

def rec_tail_time():
    
    for a in n:
        start =perf_counter_ns()

        tail_factorial(a)
        
        #results_re_tail.append(tail_factorial(a))

        end = perf_counter_ns()

        tail_t.append(end-start)
        
    return tail_t, #results_re_tail
    
rec_tail_time()

fig = plt.figure()
ax1 = fig.add_subplot()
ax1.set_ylabel('time (ns)')
ax1.set_xlabel('n values')
plt.plot(n,t, c='b', label='iterative time') 
plt.plot(n, nontail_t, c='y', label='recursive non-tail')
plt.plot(n, tail_t, c='g', label='recursive tail')
plt.legend()
plt.show()
```

### $Conclusion:$

The fact is that recursion is rarely the most efficient approach to solving a problem, and iteration is almost always more efficient. This is because there is usually more overhead associated with making recursive calls due to the fact that the call stack is so heavily used during recursion

### $Iterative:$

```python
# Iterative Binary Search Function
# It returns index of x in given array arr if present,
# else returns -1
def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    mid = 0
    
    
    while low <= high:
 
        mid = (high + low) // 2
 
        # If x is greater, ignore left half
        if arr[mid] < x:
            low = mid + 1
 
        # If x is smaller, ignore right half
        elif arr[mid] > x:
            high = mid - 1
 
        # means x is present at mid
        else:
            return mid
 
    # If we reach here, then the element was not present
    return -1
```

```python
sequence3 = [500, 1000]

binary_tb=[]
l = []

while len(sequence3) < 20:
    l.append(len(sequence3))
    target3 = sequence3[len(sequence3)//2]
    start =perf_counter_ns()
    binary_search(sequence3, target3)
    end = perf_counter_ns()
    binary_tb.append(end-start)
    
    sequence3.append(sum(sequence3))

fig, ax = plt.subplots(figsize=(16,10))

ax.set_xlabel('n')
ax.set_ylabel('time')
plt.plot(l, binary_tb, label='iter_binary_best_time')

plt.legend()

plt.show()
```

long binary search will take on input of size n, as a function of n.

At each stage of the search (pass through the body of the while loop) you split the input in half, so you successively reduce the size of the problem (h-l) this way:
n,n/2,n/4,n/8….

Clearly you will be done when the input is 1, for there's just one place. That index is the answer.

So you want the number of steps k such that $\mathcal{n}/{2}^k≤1.$ That's the smallest $k$ for which $\mathcal{2}^k≥n.$ The definition of the logarithm says that $k$ is about $\mathcal{log}_2(n)$, so binary search has that complexity.
